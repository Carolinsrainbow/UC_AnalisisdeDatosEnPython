Pregunta 1: ¿Cuál de las siguientes opciones corresponde al código que permite calcular el total de camas críticas para cada uno de los servicios de salud?

Pregunta 4
¿Cuál de las siguientes alternativas corresponde al promedio diario (redondeado con 0 decimales) de pacientes que han llegado a la red durante los últimos 10 días (desde el día 415 al día 424)?  

```python
# Creamos la tabla auxiliar donde tomamos los datos para sacar el promedio ordenados por el día de llegada (número)
aux = Info_sheet[['Día de llegada','Número de llegada']].groupby('Día de llegada').count()
```

```python
#Elegimos los últimos 10 datos (días)
aux.tail(10)
```

```python
#Sacamos el promedio de los últimos 10 datos (días)
aux.tail(10).mean()
```

```python
#Para redondear a cero
np.round(aux.tail(10).mean(),0)
```

Pregunta 6:
¿Cuál de los siguientes códigos permite calcular la cantidad de días de estadía promedio, por GRD, para los pacientes ingresados durante los últimos 10 días registrados en el archivo (y entonces, desde el día 415 al día 424)?  

```python
#La base viene con un error por lo que necesitamos crear una tabla auxiliar
Info_sheet_aux = Info_sheet.drop(columns=['Hospital al que llega'])

#Se incorpora el condicial de los días y se agrupa por GRD para sacar el promedio
Info_sheet_aux.loc[Info_sheet_aux['Día de llegada']>415].groupby('GRD').mean()
```
