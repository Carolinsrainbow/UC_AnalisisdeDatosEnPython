```python
# Filtrar los datos para el Hospital Del Salvador y GRD9
apoyo4 = Info_sheet.loc[(Info_sheet["Hospital al que llega"] == "Hospital Del Salvador") & (Info_sheet["GRD"] == "GRD9")]

# Agrupar por día de llegada y contar los pacientes
hosp_gr9 = apoyo4.groupby('Día de llegada').size().reset_index(name='Pacientes')

# Seleccionar los días de interés (420 a 424)
dias_interes = hosp_gr9[(hosp_gr9['Día de llegada'] >= 420) & (hosp_gr9['Día de llegada'] <= 424)]

# Verificar los datos seleccionados
print("Datos de los días de interés (420 a 424):")
print(dias_interes)

# Preparar los datos para la regresión
X = dias_interes[['Día de llegada']]
y = dias_interes[['Pacientes']]

regressor = LinearRegression().fit(X, y)
print(f"Pendiente: {regressor.coef_[0][0]}, Intercepto: {regressor.intercept_[0]}")

prox_llegada = 424*regressor.coef_[0][0]+regressor.intercept_[0]
print(f"Llegadas estimadas para el día 424: {np.round(prox_llegada,0)}")
```
